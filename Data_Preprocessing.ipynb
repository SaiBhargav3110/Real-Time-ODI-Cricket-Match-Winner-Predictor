In the original dataset we have two types of matches namely ,
International and test.

We are considering International matches as our problem statement is predicting match winner for international matches.

```python
international_count = data[data['team_type'] == 'international'].shape[0]

print("Number of international matches:",international_count)
```

```python
# Create a new DataFrame with only 'international' rows
international_matches = data[data['team_type'] == 'international'].copy()


international_matches.head()
```

```python
international_matches.tail()
```

```python
international_matches['match_type'].unique()
```

In international tornuments there are only 3 types of formats,
1. Test
2. ODI
3. T20

   So changing IT20 ("International T20") as T20 and ODM ("One day match") as ODI.

```python
# Replace values in the 'match_type' column
international_matches['match_type'] = international_matches['match_type'].replace({
    'ODM': 'ODI',
    'IT20': 'T20'
})

# Verify the changes
print(international_matches['match_type'].unique())
```

```python
# Count the rows where match_type is 'MDM'
mdm_count = international_matches[international_matches['match_type'] == 'MDM'].shape[0]

print("Number of records with match_type 'MDM':", mdm_count)
```

```python
# Remove rows where match_type is 'MDM'
international_matches = international_matches[international_matches['match_type'] != 'MDM']

# Verify that the rows have been deleted
print(international_matches['match_type'].unique())
```

## Column Description

1. **match_id**: Unique identifier for each match.  
2. **city**: City where the match was played.  
3. **date**: Date of the match.  
4. **player_of_match**: Name of the player who was awarded "Player of the Match".  
5. **venue**: Stadium or venue where the match was played.  
6. **match_type**: Match type (e.g., ODI, Test, T20).  
7. **team1**: Name of the first team participating in the match.  
8. **team2**: Name of the second team participating in the match.  
9. **toss_winner**: Name of the team that won the toss.  
10. **toss_decision**: Decision taken by the toss-winning team (either 'field' or 'bat').  
11. **winner**: Name of the team that won the match.  
12. **result**: The result of the match (e.g., 'runs', 'wickets', 'tie', etc.).  
13. **outcome_by**: Number of runs/wickets.  
14. **result_margin**: The margin by which the winning team won the match.  
15. **result_margin_type**: Type of margin by which the winning team won (e.g., runs or wickets).  
16. **umpire1**: Name of the first on-field umpire.  
17. **umpire2**: Name of the second on-field umpire.  
18. **season**: Year of the match.  
19. **team_type**: Indicates if the match was at the International or Club level.
